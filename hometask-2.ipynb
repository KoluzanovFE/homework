Exercise 6.

Написать функцию, реализующую алгоритм k-means. Входные и выходные значения функции должны быть аналогичны https://docs.scipy.org/doc/scipy-0.15.1/reference/generated/scipy.cluster.vq.kmeans.html. Функция должна содержать только один цикл while (for).

```python
import matplotlib.pyplot as plt
#http://www.stat.cmu.edu/~larry/all-of-statistics/=data/faithful.dat
data = np.loadtxt('1.txt', skiprows=26)[:, 1:]

plt.scatter(*data.T)
plt.xlabel('Eruption time (min)')
plt.ylabel('Waiting time ti next eruption (min)')
plt.show()
#print(data)


data = (data - np.mean(data, axis = 0))/(np.std(data, axis = 0))

plt.scatter(*data.T)
plt.show()


centroids = np.random.uniform(-2, 2, 4).reshape((2, 2))
#print(centroids, '\n')


#dist0 = np.linalg.norm(data-centroids[0], axis=1)
#dist1 = np.linalg.norm(data-centroids[1], axis=1)

#dist = np.vstack((dist0,dist1)).T
#print(dist)
#number = np.argmin(dist, axis=1)
#print(number)

def get_labels(pts, centroids):
    rasst = 100
    while (rasst > 0.000000001):
        dist0 = np.linalg.norm(pts-centroids[0], axis=1)
        dist1 = np.linalg.norm(pts-centroids[1], axis=1)
        dist = np.vstack((dist0,dist1)).T
        number = np.argmin(dist, axis=1)

        pts0=pts[ number ==0 ]
        pts1=pts[ number ==1 ]


        new_centroids=np.array([np.mean(pts0, axis=0), np.mean(pts1, axis=0)] )

        rasst = np.max([np.linalg.norm(new_centroids[0]-centroids[0]), np.linalg.norm(new_centroids[1]-centroids[1])])
        centroids=new_centroids

    return number

labels = get_labels(data, centroids)

import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(*data.T, c=np.where(labels, "green", "red"), s=20)
plt.scatter(*centroids.T, c=["red", "green"], s=95, marker='*')
plt.title('Predicted Classes')
plt.show()

```
